### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|')
users.head(5)
```

### Step 4. Discover what is the mean age per occupation

```python
users_occ_mean_age = users.groupby('occupation').mean()['age']
users_occ_mean_age
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users['gender_n'] = users['gender'].apply(lambda x : 1 if x == "M" else 0)
males_per_occ = users.groupby('occupation')['gender_n'].sum()
# when to use value_counts versus count?
total_ppl_per_occ = users.groupby('occupation').count()['gender_n']
ratio = males_per_occ/ total_ppl_per_occ * 100 
ratio.sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
# Solution 1 
max_age = users.groupby('occupation').max()['age']
min_age = users.groupby('occupation').min()['age']

```

```python
# Solution 2 
users.groupby('occupation')['age'].agg(['min','max'])

```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
# for each combination == "groupby"
users.groupby(['occupation','gender'])['age'].mean()
```

### Step 8.  For each occupation present the percentage of women and men

```python
# calling count function specifically on gender 

gender_per_occ = users.groupby(['occupation','gender']).agg({'gender': 'count'})
occ_total = users.groupby(['occupation']).agg('count')
# what is level?
gender_per = gender_per_occ.div(occ_total,level='occupation') * 100

# show all rows from gender
gender_per.loc[:,'gender']
```
